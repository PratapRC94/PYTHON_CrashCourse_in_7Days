# See them in Action

```python
import math

def addition(num_1=0,num_2=0):
    """
    Function --> add two numbers
    Arguments:
        x --> float (default 0.0)
        y --> float (default 0.0)
    Return Type --> float
    """
    
    # TRY BLOCK
    try:
        return float(num_1) + float(num_2)
    
    #EXCPET BLOCK
    except TypeError as te:
        return 'Type error. Please check Input Type'
    except ValueError as ve:
        return 'Value Error. Please check boundaries'
    except Exception as e: #
        return 'Unknown Exception occured'
```

```python
a, b = (1, 2), (6, 7)
addition(a, b)
```

```python
a, b = "a", 6
addition(a, b)
```

```python
a, b = 4, 6
addition(a, b)
```

```python
def addition(num_1=0,num_2=0):
    """
    Function --> add two numbers
    Arguments:
        x --> float (default 0.0)
        y --> float (default 0.0)
    """
    
    # TRY BLOCK
    try:
        print("Sum of 2 input numbers : ", float(num_1) + float(num_2) )
    
    # EXCPET BLOCK
    except TypeError as te:
        print( 'Type error. Please check Input Type' )
    except ValueError as ve:
        print( 'Value Error. Please check boundaries' )
    except Exception as e:
        print( 'Unknown Exception occured' )
    
    # Else Block
    else:
        print( "Exceuted without any error" )
        
    #FINALLY BLOCK
    finally:
        print( "Finally block executed." )
```

```python
a, b = 4, "d"
addition(a, b)
```

```python
a, b = 4, 6
addition(a, b)
```

```python
def add(a, b):
    try:
        print(a+b)
    except:
        print("exception ocured")
```

```python
def subtraction(num_1,num_2):
    """
    Function --> subtract two numbers
    Arguments:
        x --> float (default 0.0)
        y --> float (default 0.0)
    Return Type --> float
    """
    try:
        return float(num_1)-float(num_2)
    except TypeError as te:
        return 'Type error. Please check Input Type'
    except ValueError as ve:
        return 'Value Error. Please check boundaries'
    except Exception as e:
        return 'Unknown Exception occured'

def multiplication(num_1,num_2):
    """
    Function --> multiply two numbers
    Arguments:
        x --> float (default 0.0)
        y --> float (default 0.0)
    Return Type --> float
    """
    try:
        return float(num_1)*float(num_2)
    except TypeError as te:
        return 'Type error. Please check Input Type'
    except ValueError as ve:
        return 'Value Error. Please check boundaries'
    except Exception as e:
        return 'Unknown Exception occured'
        
def division(num_1,num_2):
    """
    Function --> divide two numbers
    Arguments:
        x --> float (default 0.0)
        y --> float (default 0.0)
    Return Type --> float
    """
    try:
        return float(num_1)/float(num_2)
    except ZeroDivisionError as ze:
        return 'Cannot divide by 0'
    except TypeError as te:
        return 'Type error. Please check Input Type'
    except ValueError as ve:
        return 'Value Error. Please check boundaries'
    except Exception as e:
        return 'Unknown Exception occured'
```

# Custom Exceptions (User Defined)

```python
# define Python user-defined exceptions
class Error(Exception):
    """Base class for other exceptions"""
    pass
class ValueTooSmallError(Error):
    """Raised when the input value is too small"""
    pass
class ValueTooLargeError(Error):
    """Raised when the input value is too large"""
    pass

# you need to guess this number
number = 10

# user guesses a number until he/she gets it right
while True:
    try:
        i_num = int(input("Enter a number: "))
        if i_num < number:
            raise ValueTooSmallError
        elif i_num > number:
            raise ValueTooLargeError
        break
    except ValueTooSmallError:
        print("This value is too small, try again!")
        print()
    except ValueTooLargeError:
        print("This value is too large, try again!")
        print()

print("Congratulations! You guessed it correctly.")
```

## Difference between Exception & Assertion
Exception is a mechanism of checking if the implementation is executing without any expected or unexpected errors or not. So, we see that exceptions are basically used for handling even the unforseen conditions during the execution of an application in a better way and hence using exceptions effectively results into a robust application.

Assertions should never be a part of the implementation of some functionality of the application. They should only be used to verify the assumptions - just to be sure that whatever we assumed while desinging the solution is actually valid in practical as well.

### fibonacci series

```python
def fib(n):
    l=[]
    a=0
    b=1
    count=0
    if n==1:
        print('fibonacci series upto {n} terms: ',a)
    else:
        
        while count<n:
            l.append(a)
            x=a+b
            a=b
            b=x
            count+=1
            
    return(l)
    
res=fib(10)
print(res)
```

```python
li=[]
def recur_fibo(n):
    if n <= 1:
        return n
    else:
        return(recur_fibo(n-1) + recur_fibo(n-2)) 

#nterms = 10
nterms = int(input("no of terms? :  "))

# check if the number of terms is valid
#if nterms <= 0:
   #print("Plese enter a positive integer")
#else:
print("Fibonacci sequence:")
for i in range(nterms):
    li.append(recur_fibo(i))
print(li)
```
